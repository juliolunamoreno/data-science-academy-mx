```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```
def game(dice_max = 6, goal = 100):
    pos = 0
    turns = 0
    SERPIENTES_ESCALERAS = {10:38, 4:14, 17:31, 16:6, 21:42, 28:84, 36:1,
                  47:26, 49:11, 51:67, 56:53, 62:19, 64:60,
                  71:91, 80:100, 87:24, 93:73, 95:75, 98:78}
    while pos < goal:
        dice = np.random.randint(1,dice_max + 1)
        pos += dice

        if pos > goal:
            pos = goal -(pos + dice - goal)

        pos = SERPIENTES_ESCALERAS.get(pos, pos)
        turns+=1
    return turns
```

```
game(10)
```

```
games = 2000
ls = [game() for _ in range(games)]

fig, ax = plt.subplots(1,2, figsize = (12,5))
ax[0].hist(ls, color = 'steelblue', edgecolor = 'black', bins = 'auto')
ax[0].set_xlabel('Number of turns')
ax[0].set_ylabel('Number of games finished')
ax[0].set_title('Turns distribution')

ax[1].hist(ls, color = 'steelblue', cumulative = True, alpha = .9, edgecolor = 'black')
ax[1].set_xlabel('Number of turns')
ax[1].set_ylabel('Fraction of games finished')
ax[1].set_title('Turns cumulative distribution');
```
